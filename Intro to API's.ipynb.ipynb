```python
import requests

def get_pokemon_data(pokemon_id):
    url = f'https://pokeapi.co/api/v2/pokemon/{pokemon_id}/'
    response = requests.get(url)

    if response.status_code == 200:
        return response.json()
    else:
        print(f"Error: Unable to retrieve data for Pokemon with ID {pokemon_id}")
        return None

def create_pokemon_dictionary(pokemon_data):
    name = pokemon_data['name']
    abilities = [ability['ability']['name'] for ability in pokemon_data['abilities']]
    weight = pokemon_data['weight']

    return {name: {'abilities': abilities, 'weight': weight}}

def organize_pokemon_by_type(pokemon_list):
    organized_dict = {}

    for pokemon_data in pokemon_list:
        name = pokemon_data['name']
        types = pokemon_data['types']

        for pokemon_type in types:
            type_name = pokemon_type['type']['name']

            if type_name not in organized_dict:
                organized_dict[type_name] = {}

            organized_dict[type_name].update(create_pokemon_dictionary(pokemon_data))

    return organized_dict

def main():
    pokemon_list = []

    # Retrieve data for 20 Pokemon
    for pokemon_id in range(1, 21):
        pokemon_data = get_pokemon_data(pokemon_id)
        if pokemon_data:
            pokemon_list.append(pokemon_data)

    # Organize Pokemon by type
    organized_pokemon = organize_pokemon_by_type(pokemon_list)

    # Print the result
    print(organized_pokemon)

if __name__ == "__main__":
    main()
```
